```python
nb['win']=nb["果"].map({"胜": 1, "负": 0})
```

```python
winning=nb.groupby(["赛季"])["win"].mean()[season].values
nb2=nb[['赛季','win']]
sjsc=nb2.groupby('赛季').sum()  #赛季胜场
nb['统计']=nb["果"].map({"胜": 1, "负": 1})
nb3=nb[['赛季','统计']]
sjc=nb3.groupby('赛季').sum()   #赛季场
dfl=sjsc.join(sjc, how='inner')  #得分率
dfl['胜率']=dfl['win']/dfl['统计']
```
